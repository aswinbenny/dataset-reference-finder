```python
# Define refined patterns again
refined_patterns = {
    'SAMN BioSample': r'^SAMN\d+$',
    'EPI_ISL (GISAID)': r'^EPI_ISL_\d+$',
    'CHEMBL': r'^CHEMBL\d+$',
    'IPR': r'^IPR\d{6}$',
    'ENSEMBL Gene ENSBTAG': r'^ENSBTAG\d{11}$',
    'ENSEMBL Gene ENSOARG': r'^ENSOARG\d{11}$',
    'ENSEMBL Gene ENSMMUT': r'^ENSMMUT\d{11}$',
    'STH ID': r'^STH\d{5}$',
    'Short Alphanumeric ID (PDB-like)': r'^[0-9][A-Z0-9]{3}$',
    'PF': r'^PF\d{5}$',
    'PRJNA (BioProject)': r'^PRJNA\d+$',
    'E-GEOD (GEO)': r'^E-GEOD-\d+$',
    'KEGG K Numbers': r'^K\d{5}$',
    'CVCL (Cellosaurus)': r'^CVCL_\w+$',
    'UniProt ID': r'^[OPQ][0-9][A-Z0-9]{3}[0-9]$',
    'EPI Short (Legacy)': r'^EPI\d+$',
    'E-PROT (Expression Atlas)': r'^E-PROT-\d+$',
    'PXD (ProteomeXchange)': r'^PXD\d{6,}$',
    'HPA ID': r'^HPA\d+$',
    'SRX/SRR/SRP': r'^SR[PXR]\d+$',
    'EMPIAR': r'^EMPIAR-\d+$',
    'GSE': r'^GSE\d+$',
    'CP GenBank/RefSeq': r'^CP\d+(\.\d+)?$',
    'EC Numbers': r'^\d+\.\d+\.\d+\.\d+$',
    'PDB ID': r'^[1-9][a-z0-9]{3}$',
    'CAB ID': r'^CAB\d+$',
    'rs ID': r'^rs\d+$',
    'NM_ RefSeq': r'^NM_\d+$',
    'NC_ RefSeq': r'^NC_\d+\.\d+$',
    'BX ID': r'^BX\d+$',
    'D ID': r'^D\d+$',
    'KX ID': r'^KX\d+$',
    'ERR ID (ENA Run)': r'^ERR\d+$',
    'MODEL ID': r'^MODEL\d+$',
    'F ID': r'^F\d+[A-Z0-9]+$',
}

# Match against refined patterns
pattern_matches = defaultdict(list)
for identifier in ids:
    matched = False
    for name, pattern in refined_patterns.items():
        if re.fullmatch(pattern, identifier):
            pattern_matches[name].append(identifier)
            matched = True
            break
    if not matched:
        pattern_matches['Unmatched'].append(identifier)

# Build summary table
pattern_summary = [
    {
        "Pattern": pattern,
        "Matching Words": matches,
        "Count": len(matches)
    }
    for pattern, matches in sorted(pattern_matches.items(), key=lambda x: -len(x[1]))
]


# Display the summary table
import pandas as pd
summary_df = pd.DataFrame(pattern_summary)
summary_df
```
